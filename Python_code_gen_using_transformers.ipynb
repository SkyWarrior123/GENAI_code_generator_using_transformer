```python
# importing essential libraries

import torch
from transformers import pipeline

# handling randomness
def set_seed(seed):
  torch.manual_seed(seed)
  if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)
```

```python
# Loading th efinetuned model
pipe = pipeline("text-generation", model="GuillenLuis03/PyCodeGPT")
```

```python
# Code Gen 1
propmt = "Give me a function to reverse a string"
code_gen  = pipe(propmt,
                 max_length=100,
                 temperature=0.7,
                 num_return_sequences=1
                 )[0]['generated_text']
print("Here's your python code below------------------------------------")
print(code_gen)
```

```python
propmt = "Give me function to add, subtract, divide and multiply two numbers"

code_gen  = pipe(propmt,
                 max_length=1000,
                 temperature=0.7,
                 num_return_sequences=1
                 )[0]['generated_text']
print("Here's your python code below------------------------------------")
print(code_gen)
```
